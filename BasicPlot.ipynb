```python
#importera bibliotek
import matplotlib.pyplot as plt
```

```python
#skapa listor för familjemedlemmar och ålder (två separata listor)
familjeNamn = ['Anna','Zoran','Ines','Gunnar','Kent']
namnAlder = [45,43,7,13,30]
```

```python
#(piffa upp plot med beskrivande text på axlarna)
#('trimma' axlarna till lämplig skala)
#plotta!
plt.bar(familjeNamn,namnAlder)
plt.ylim(0,20)
plt.title('Familjen Svenssons Ålder')
plt.ylabel('Ålder (år)')
plt.savefig('plot.png')
plt.show()
```
