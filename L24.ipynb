```python
t = np.arange(0,3,1e-3)
plt.plot(t)
```

```python
T = 1
f = 1/T
sinValues = 0.5 * np.sin(2 * np.pi * f * t)
```

```python
plt.plot(t,sinValues)
```

```python
def Bn(amp, n):
    #om n udda - returner beräknat värde
    if n % 2 != 0:
        return (4 * amp) / (n * np.pi)
    #...annars returenera 0
    else:
        return 0
```

```python
numOfKoeff = 600
fourierKoeff = []
for i in range(0,numOfKoeff):
    fourierKoeff.append(Bn(0.5,i))
    
fourierSum = [0] * 3000
#itterera över fourierKoeff och addera var sinus kurva
for i in range(0,numOfKoeff):
    fourierSum += fourierKoeff[i] * np.sin(i * 2 * np.pi * f * t)
    
plt.plot(t,fourierSum)
```

```python
plt.plot(t,np.sin(2*np.pi*400*t))
plt.xlim(0,0.01)
plt.show()
```
